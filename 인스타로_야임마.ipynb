```python
# ==============================================================================
# [1단계] 크롤링에 필요한 도구(패키지) 가져오기
# ==============================================================================

# selenium 라이브러리에서 웹 브라우저를 제어하는 핵심 기능인 webdriver를 가져옵니다.
from selenium import webdriver

# 브라우저를 켜고 끄는 등의 서비스를 관리하는 Service 도구를 가져옵니다.
from selenium.webdriver.chrome.service import Service

# 웹 페이지의 요소(버튼, 입력창 등)를 '어떤 방식'으로 찾을지 결정하는 By 도구를 가져옵니다.
# (예: ID로 찾기, 클래스 이름으로 찾기, CSS 선택자로 찾기 등)
from selenium.webdriver.common.by import By

# 키보드의 엔터, 백스페이스, 화살표 같은 특수 키를 입력할 수 있게 해주는 Keys를 가져옵니다.
from selenium.webdriver.common.keys import Keys

# 컴퓨터에게 "잠시 대기해라"라는 명령을 내리기 위해 파이썬 내장 라이브러리인 time을 가져옵니다.
# (웹 페이지가 로딩되는 시간을 기다려주지 않으면 에러가 나기 때문입니다.)
import time


# ==============================================================================
# [2단계] 크롤링 제어 및 자동화 실행
# ==============================================================================

# 가상의 크롬 브라우저를 실행합니다. (실행 시 빈 크롬 창이 새로 뜨게 됩니다.)
# 최신 셀레니움은 별도의 드라이버 다운로드 없이 이 한 줄로 자동 실행됩니다.
driver = webdriver.Chrome()

try:
    # driver야, 괄호 안의 인터넷 주소(네이버)로 접속해줘!
    driver.get("https://www.naver.com")
    
    # 웹 페이지가 완전히 켜지고 안정화될 때까지 2초 동안 가만히 대기합니다.
    time.sleep(2)
    
    # 네이버 메인 화면에서 id 속성 값이 "query"인 요소(검색창 입력칸)를 찾아서 search_box에 저장합니다.
    search_box = driver.find_element(By.ID, "query")
    
    # 찾은 검색창에 "웹 크롤링"이라는 글자를 타이핑하듯이 입력합니다.
    search_box.send_keys("웹 크롤링")
    
    # 글자를 다 쳤으니 키보드의 엔터(ENTER) 키를 누르는 신호를 보냅니다.
    search_box.send_keys(Keys.ENTER)
    
    # 검색 결과 페이지가 로딩되고, 사람이 눈으로 확인할 수 있도록 5초 동안 화면을 멈춥니다.
    time.sleep(5)

finally:
    # 코드가 성공하든, 중간에 에러가 나든 관계없이 '무조건' 마지막에는 실행되는 블록입니다.
    # 열려있는 크롬 브라우저 창을 닫고 프로그램을 안전하게 종료합니다.
    # (이 코드가 없으면 컴퓨터 메모리에 크롬 창이 계속 남아있게 됩니다.)
    driver.quit()
```
