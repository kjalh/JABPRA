```python
# 첫과 끝 정하지 않고 슬라이싱으로 대체

f = int(input("찾는 수 입력: "))
arr = [1, 2, 3, 4, 5, 6, 7]

c_arr = arr.copy()

while True:
    if len(arr) == 0:
        print("찾는 수 없음")
        break
        
    mid = len(arr) // 2
    
    if f == arr[mid]:
        print(f"{f}는 {c_arr.index(f) + 1}번째에 있음")
        break
    elif f < arr[mid]:
        arr[:] = arr[0: mid]
    elif f > arr[mid]:
        arr[:] = arr[mid + 1:]
        
```
